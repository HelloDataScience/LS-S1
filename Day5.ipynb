```python
# 강남구 법정동별 매매건수를 세로 막대 그래프로 그려보세요. 그리고 대치동을 빨간색으로 채워보세요.
# 막대 위에 매매건수를 텍스트로 추가해야 합니다.
cond = apt['시군구'].eq('강남구')
gng = apt[cond]
grp = gng['법정동'].value_counts()
mypal = ['silver'] * gng['법정동'].nunique()
mypal[2] = 'red'
sns.countplot(data = gng, x = '법정동', color = '0.8', order = grp.index, 
              hue = '법정동', palette = mypal, hue_order = grp.index)
plt.xlim(-1, 14)
plt.ylim(0, 2250)
for i, v in enumerate(grp):
    plt.text(x = i, y = v + 20, s = f'{v:,}', ha = 'center', va = 'bottom');
```

```python
sns.barplot(data = apt, x = '시군구', y = '거래금액', edgecolor = '0.5', 
            estimator = 'mean', errorbar = None, order = grp.index, 
            hue = '시군구', palette = mypal.tolist(), hue_order = grp.index)
plt.xlim(-1, 25)
plt.ylim(0, 25)
plt.xticks(rotation = 90, fontsize = 7)
for i, v in enumerate(grp):
    plt.text(x = i, y = v + 0.2, s = f'{v:.1f}억', ha = 'center', va = 'bottom', 
             size = 7, color = '0');
```

```python
plt.figure(figsize = (8, 6))
grp = gng.groupby(by = '법정동')['평당금액'].mean().sort_values(ascending = False)
mypal = pd.Series(data = ['white'] * gng['법정동'].nunique())
idx = np.where(grp.ge(8000))[0]
mypal[idx] = 'gold'
sns.barplot(data = gng, x = '평당금액', y = '법정동', edgecolor = '0.5', 
            estimator = 'mean', errorbar = None, order = grp.index, 
            hue = '법정동', palette = mypal.tolist(), hue_order = grp.index)
plt.xlim(0, 12000)
plt.ylim(14, -1)
for i, v in enumerate(grp):
    plt.text(x = v - 100, y = i, s = f'{int(round(v, 0)):,}만원', ha = 'right', va = 'center', 
             size = 7, color = '0', fontweight = 'bold');
```
