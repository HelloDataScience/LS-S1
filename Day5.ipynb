```python
cond = apt['시군구'].str.contains(pat = '강[남동북]')
sub = apt.loc[cond, :]
```

```python
pd.pivot_table(data = sub, index = '시군구', values = '거래금액', 
               aggfunc = ['count', 'mean', 'std'])
```

```python
sns.histplot(data = sub, x = '거래금액', binwidth = 2, binrange = (0, 70), hue = '시군구');
```

```python
sns.kdeplot(data = sub, x = '거래금액', hue = '시군구', fill = True)
plt.xlim(-4, 84)
plt.title(label = '시군구별 아파트 거래금액 분포 비교', 
          fontdict = dict(fontweight = 'bold', size = 16, color = 'blue'))
plt.xlabel(xlabel = '거래금액(억)')
plt.ylabel(ylabel = '밀도');
```

```python
out_dict = {
    'marker': '*', 
    'markersize': 3, 
    'markerfacecolor': 'pink', 
    'markeredgecolor': 'red', 
    'markeredgewidth': 0.5
}
```

```python
sns.boxplot(data = apt, y = '거래금액', color = '0.8', 
            linewidth = 0.5, flierprops = out_dict);
```

```python
grp = apt.groupby(by = '시군구')['거래금액'].median().sort_values()
grp
```

```python
sns.boxplot(data = apt, x = '시군구', y = '거래금액', color = '0.8', 
            linewidth = 0.5, flierprops = out_dict, order = grp.index)
plt.axhline(y = apt['거래금액'].median(), color = 'red', lw = 1, ls = '--')
plt.axvline(x = 9.5, color = 'blue')
plt.axvline(x = 16.5, color = 'blue')
plt.xticks(rotation = 90, fontsize = 7)
plt.xlim(-1, 25);
```

```python
grp = apt['시군구'].value_counts()
grp
```

```python
mypal = ['red'] + ['silver'] * 24
```

```python
sns.countplot(data = apt, x = '시군구', color = '0.8', order = grp.index, 
              hue = '시군구', palette = mypal, hue_order = grp.index)
plt.xlim(-1, 25)
plt.ylim(0, 22500)
plt.xticks(rotation = 90, fontsize = 7)
for i, v in enumerate(grp):
    plt.text(x = i, y = v + 200, s = f'{v:,}', ha = 'center', va = 'bottom', 
             size = 7, color = 'blue');
```

```python
# 강남구 법정동별 매매건수를 세로 막대 그래프로 그려보세요. 그리고 대치동을 빨간색으로 채워보세요.
# 막대 위에 매매건수를 텍스트로 추가해야 합니다.
cond = apt['시군구'].eq('강남구')
gng = apt[cond]
grp = gng['법정동'].value_counts()
mypal = ['silver'] * gng['법정동'].nunique()
mypal[2] = 'red'
sns.countplot(data = gng, x = '법정동', color = '0.8', order = grp.index, 
              hue = '법정동', palette = mypal, hue_order = grp.index)
plt.xlim(-1, 14)
plt.ylim(0, 2250)
for i, v in enumerate(grp):
    plt.text(x = i, y = v + 20, s = f'{v:,}', ha = 'center', va = 'bottom');
```
