```python
import requests
import xmltodict
import pandas as pd
```

```python
def apt_price(areaCd, ymonth, page = 1, rows = 9999):
    url = 'http://apis.data.go.kr/1613000/RTMSDataSvcAptTrade/getRTMSDataSvcAptTrade'
    query = {
        'LAWD_CD': areaCd, 
        'DEAL_YMD': ymonth, 
        'serviceKey': '4zNXftfpIAhhRRJLlpBmH8qMqYRZBgNxXWDoHow+WR1Di3VuLplaDdiTX/X3pooreCQBuT4dtCX1ENw8EX4x/Q==',
        'pageNo': page, 
        'numOfRows': rows
    }
    res = requests.get(url = url, params = query)
    dat = xmltodict.parse(xml_input = res.text)
    df = pd.DataFrame(data = dat['response']['body']['items']['item'])
    return df
```

```python
df = apt_price(areaCd = '11680', ymonth = '202504')
```
